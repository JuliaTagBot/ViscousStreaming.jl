```julia
Re = 40
α = im*Re+1
β = sqrt(α)
γ₁ = β + conj(β)
γ₂ = conj(β)+1
C₁ = 0.5*Re*(conj(β)+1)*(β-conj(β))/(β-1)
C₂ = -0.5*Re*β*(1-conj(β))*(1+conj(β))/(β-1)
Ĉ₁ = C₁/(γ₁^2-4)^2
Ĉ₂ = C₂/(γ₂^2-4)^2
```

```julia
ψ₁(x,y,t) = real((β*exp(-y) - exp(-β*y))/(β-1)*cos(x)*exp(im*t))
```

```julia
ψ̄₂(x,y) = -0.5*real(((-Ĉ₁-Ĉ₂-0.5*y*((γ₁-2)Ĉ₁+(γ₂-2)Ĉ₂))*exp(-2y)+Ĉ₁*exp(-γ₁*y)+Ĉ₂*exp(-γ₂*y))*sin(2x))
```

```julia
ũ₁(y) = β/(β-1)*(exp(-β*y)-exp(-y))
ṽ₁(y) = (β*exp(-y)-exp(-β*y))/(β-1)
```

```julia
ψ̄d(x,y) = -0.25*sin(2*x)*real(im*(ũ₁(y))*conj(ṽ₁(y)))
```
